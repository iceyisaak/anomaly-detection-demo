# DBScan Anomaly Detection - DEMO

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Dataset

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=750, factor=0.3,noise=0.1, random_state=42)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

---

## DBScan

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.1)
```

```python
dbscan.fit_predict(X)
```

```python
pred = dbscan.labels_
```

```python
pred
```

```python
outlier_index = np.where(pred < 0)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X[outlier_index,0],X[outlier_index,1],c="red")
plt.title('Dataset with Highlighted Outliers')

plt.show();
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Complete Dataset")

plt.show();
```

```python
plt.scatter(X[:,0],X[:,1],c=pred)
plt.title('Predicted Cluster with Outliers')

plt.show();
```
