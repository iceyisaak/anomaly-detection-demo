# Isolation Forest - Anomaly Detection DEMO

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

# Dataset

```python
from sklearn.datasets import make_blobs, make_classification
```

```python
X,y = make_blobs(n_samples=500, n_features=2, centers=2, random_state=42, cluster_std=0.8)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

---

## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_for = IsolationForest(contamination=0.2)
```

```python
iso_for.fit(X)
```

```python
pred = iso_for.predict(X)
```

```python
pred
```

```python
outlier_index = np.where(pred < 0)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(X[outlier_index,0],X[outlier_index,1],c="red");
```

```python
fig = plt.figure(figsize=(15,15))


ax1 = fig.add_subplot(2,2,1)
ax1.scatter(X[:,0],X[:,1])
ax1.set_title('All Data Points')
ax1.set_xlabel('Feature 1')
ax1.set_ylabel('Feature 2')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(X[outlier_index,0],X[outlier_index,1],c="red", label='Outliers')
ax2.set_title('Outliers Only ')
ax2.set_xlabel('Feature 1')
ax2.set_ylabel('Feature 2')

ax3 = fig.add_subplot(2,1,2)
ax3.scatter(X[:, 0], X[:, 1], c='blue', alpha=0.6, label='Inliers')
ax3.scatter(X[outlier_index, 0], X[outlier_index, 1], c="red", s=50, label='Outliers')
ax3.set_title('All Data with Outliers Highlighted')
ax3.set_xlabel('Feature 1')
ax3.set_ylabel('Feature 2')
ax3.legend()


plt.tight_layout()
plt.show();
```
